```python
import pandas as pd 
import sklearn
from sklearn.datasets import load_breast_cancer
```

```python
#create a variable and load in a breast cancer dataset already installed in sklearn
Cancer_data = load_breast_cancer()
print(Cancer_data)
```

```python
#print of a list of feature names and target names to see what we have
print(Cancer_data.feature_names)
print(Cancer_data.target_names)
```

```python
#create a data frame with the columns as the feature names of the breast cancer dataset
df = pd.DataFrame(Cancer_data, columns= Cancer_data.feature_names)
df.head()
```

```python
#using .data.shape will tell us how many rows (569) and how many columns our dataset includes
Cancer_data.data.shape
```

```python
#It's at this point i realised i need to be .data attached to the cancer_data variable for my rows to fill in to create a dataframe
#the .data is important as it specifically requests for the numpy arrays of the Cancer_data as originally it is a bunch object
df = pd.DataFrame(Cancer_data.data, columns=Cancer_data.feature_names)
df.head()
```

```python
df["target"] = Cancer_data.target
df.head()
```

```python
Cancer_data.target_names
```

```python
df[df.target == 1].head()
```

```python
df["Cancer type"] = df.target.apply(lambda x: Cancer_data.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
```

```python
plt.xlabel("mean radius")
plt.ylabel("mean texture")
plt.scatter(df0["mean radius"], df0["mean texture"], color = "green", marker= "+")
plt.scatter(df1["mean radius"], df1["mean texture"], color = "blue", marker= "x")
```

```python
plt.xlabel("mean perimeter")
plt.ylabel("mean area")
plt.scatter(df0["mean perimeter"], df0["mean area"], color = "green", marker= "+")
plt.scatter(df1["mean perimeter"], df1["mean area"], color = "blue", marker= "x")
```

```python
from sklearn.model_selection import train_test_split
```
